```python
 import pandas as pd
import numpy as np

# Adjusted monthly weights
monthly_weights = {
    "January": 9.27,
    "February": 8.88,
    "March": 8.49,
    "April": 7.72,
    "May": 7.34,
    "June": 7.72,
    "July": 8.10,
    "August": 8.10,
    "September": 7.72,
    "October": 8.49,
    "November": 8.88,
    "December": 9.27
}

# Normalize weights to sum to 1
normalized_weights = np.array(list(monthly_weights.values())) / sum(monthly_weights.values())

# Map months to indices (0-based for numpy)
month_map = {month: idx for idx, month in enumerate(monthly_weights.keys())}

# Simplified function to process data
def yearly_to_simple_weekly(data, weights):
    weekly_data = []

    for _, row in data.iterrows():
        year = int(row['Datum'])
        yearly_total = row['Konsum(GwH)']

        # Step 1: Calculate monthly totals using weights
        monthly_totals = yearly_total * weights

        # Step 2: Divide each month's total into 4 equal weeks
        for month_idx, monthly_total in enumerate(monthly_totals):
            weekly_total = monthly_total / 4
            for week in range(4):
                week_date = pd.Timestamp(year=year, month=month_idx + 1, day=1) + pd.Timedelta(days=week * 7)
                weekly_data.append({
                    'Date': week_date,
                    'Konsum(GwH)': weekly_total
                })

    return pd.DataFrame(weekly_data)

# Transform wallisyearlyenergy into weekly data
weights_array = normalized_weights  # Normalize weights
weekly_wallisenergy = yearly_to_simple_weekly(wallisyearlyenergy, weights_array)

# Merge with merged_df_sion
weekly_wallisenergy['Date'] = pd.to_datetime(weekly_wallisenergy['Date'])
merged_df_sion['Date'] = pd.to_datetime(merged_df_sion['Date'])

# Merge weekly data into merged_df_sion
wallisdataset = pd.merge_asof(
    merged_df_sion.sort_values('Date'),
    weekly_wallisenergy.sort_values('Date'),
    on='Date',
    direction='backward'
)

wallisdataset
```
